# Single thread benchmark


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "single_thread.csv"
data = pd.read_csv(file)

image_ids = data["image"].unique()

for image_id in image_ids:
    data_image = data[data["image"] == image_id]
    image_resolution = data_image["resolution"].unique()[0]

    fig, ax = plt.subplots()

    num_points = len(data_image)
    colors = plt.cm.viridis(np.linspace(0, 1, num_points))

    ax.plot(data_image["number_of_pixel_groups"], data_image["time"], label=image_id)

    for i in range(num_points):
        ax.scatter(
            data_image["number_of_pixel_groups"].iloc[i],
            data_image["time"].iloc[i],
            color=colors[i],
            label=f'No. of pixel groups: {data_image["number_of_pixel_groups"].iloc[i]}',
        )

    ax.set_xlabel("Number of pixel groups")
    ax.set_ylabel("Time (s)")
    ax.set_title(
        f"Image {image_id} ({image_resolution}) - Time vs Number of pixel groups (single thread)"
    )
    ax.legend()

    plt.show()
```
